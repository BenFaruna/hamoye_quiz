```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('data/energydata_complete.csv', parse_dates=[0])
df.head(2)
```

```python
df.info()
```

```python
df.describe()
```

```python
df.corr()
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(), cmap=plt.cm.bwr)
```

```python
plt.rcParams.update({'figure.figsize':(20, 20), 'figure.dpi':100})
```

```python
fig, ax = plt.subplots(3, 2)
ax[0][0].plot('date', 'Press_mm_hg', 'r', data=df,)
ax[0][1].plot('date', 'T_out', 'g', data=df)
ax[1][0].plot('date', 'RH_out', 'b', data=df)
ax[1][1].plot('date', 'Windspeed', 'y', data=df)
ax[2][0].plot('date', 'Tdewpoint', 'k', data=df)
ax[2][1].plot('date', 'Visibility', 'c', data=df)
fig.legend(loc='upper right')
```
